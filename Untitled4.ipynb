```python
df = pd.read_csv("CSV.csv")
```

```python
df.head(2)
```

```python
# prompt: grafica los datos de x magnetic field contra tiempo

plt.plot(df['Data Set 1:Time(s)'], df['Data Set 1:X magnetic field(mT)'])
plt.xlabel("tiempo")
plt.ylabel("x magnetic field")
plt.title("Gráfica de x magnetic field contra tiempo")
plt.show()
```

```python
# prompt: haz una regresion polinomica de orden 2 a Data Set 1:X magnetic field(mT)

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly_features1 = PolynomialFeatures(degree=2)

# Transform the 'Data Set 1:Time(s)' feature into a polynomial feature matrix.
X_poly1 = poly_features1.fit_transform(df['Data Set 1:Time(s)'].values.reshape(-1, 1))

# Create a linear regression model.
model1 = LinearRegression()

# Fit the model to the transformed data.
model1.fit(X_poly1, df['Data Set 1:X magnetic field(mT)'])

# Predict the values using the model.
y_pred1 = model1.predict(X_poly1)

# Plot the original data and the predicted values.
plt.plot(df['Data Set 1:Time(s)'], df['Data Set 1:X magnetic field(mT)'], label="Original data")
plt.plot(df['Data Set 1:Time(s)'], y_pred1, color='red', label="Predicted values")

plt.xlabel("tiempo")
plt.ylabel("x magnetic field")
plt.title("Polynomial Regression of Order 2 for Data Set 1:X magnetic field(mT)")
plt.legend()
plt.show()
```

```python

df['x_ajust'] = y_pred1
```

```python
# prompt: grafica x_ajust contra tiempo

plt.plot(df['Data Set 1:Time(s)'], df['x_ajust'])
plt.xlabel("tiempo")
plt.ylabel("x_ajust")
plt.title("Gráfica de x_ajust contra tiempo")
plt.show()
```

```python
# prompt: borra todas las columnas menos la del tiempo y la del x_ajust

df = df[['Data Set 1:Time(s)', 'x_ajust']]
```

```python
df
```

```python
# prompt: Mediante el marco de datos df: ahora crea una nueva columna en donde se dupliquen los datos de x_ajust

df['x_ajust_duplicado'] = df['x_ajust'].copy()
```

```python
df
```

```python
# prompt: Mediante el marco de datos df: organiza los datos de x_ajust_duplicado de menor a mayor sin cambiar el orden de los datos de x_ajust. Solo quiero modificar x_ajust_duplicado

df['x_ajust_duplicado'] = df['x_ajust_duplicado'].sort_values(ignore_index=True)
```

```python
df
```

```python
# prompt: Mediante el marco de datos df: ahora en un df2 copia todos los datos de x_ajust_duplicado y x_ajust en una sola columna, primero todos los datos de x_ajust_duplicado y despues todos los datos de x_ajust

df1 = pd.DataFrame({
    'x_ajust_duplicado': df['x_ajust_duplicado'].tolist() + df['x_ajust'].tolist()
})
```

```python
df1
```

```python
# prompt: Mediante el marco de datos df2: grafrica df2

import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 7))
plt.plot(df1['x_ajust_duplicado'])
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Grafica de df2')
plt.show()
```

```python
# prompt: en la mitad del index agrega 100 index nuevos con valor de campo igual a 1.083999 	 y graficalo, reinicia los index

# Create a new dataframe with 100 rows and a constant value of 1.083999
new_df1 = pd.DataFrame({'x_ajust_duplicado': [1.083999] * 100})

# Insert the new dataframe into the original dataframe at the middle index
df1 = pd.concat([df1.iloc[:df1.shape[0] // 2], new_df1, df1.iloc[df1.shape[0] // 2:]]).reset_index(drop=True)

# Plot the modified dataframe
plt.figure(figsize=(15, 7))
plt.plot(df1['x_ajust_duplicado'])
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Modified Grafica de df1')
plt.show()
```

```python
# prompt: ponle de titulo al anterior grafico campo magnetico a partir del desplazamiento en x cambia el nombre del eje coordenado x por desplazamiento en x

plt.figure(figsize=(15, 7))
plt.plot(df1['x_ajust_duplicado'])
plt.xlabel('Desplazamiento en x')
plt.ylabel('Campo magnético')
plt.title('Campo magnético a partir del desplazamiento en x')
plt.show()
```

```python
# prompt: ahora genera una nueva columna de desplazamiento en , en donde empieza de -1m a 1m y graficala en x

df1['Desplazamiento en x'] = np.linspace(-1, 1, len(df1))

plt.figure(figsize=(15, 7))
plt.plot(df1['Desplazamiento en x'], df1['x_ajust_duplicado'])
plt.xlabel('Desplazamiento en x (m)')
plt.ylabel('Campo magnético (mT)')
plt.title('Campo magnético en función del desplazamiento en x, centro')
plt.show()
```
